```python
#defining X as the variables that are not individual data like IP address or target IP address
X_new = raw1.iloc[:, 5:42]

#using binary outcome instead of the multiclass outcome    
y_new = raw1['label']
```

```python
#removing features with low variance
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold = (.8*(1-.8)))
X = sel.fit_transform(X_new)
X.shape
```

```python
#seeing what the indices are for the low variance features
lowvar_features=sel.get_support(indices=True)
lowvar_features
```

```python
# Making training and testing set
# from finaldataframe, cut into training and testing
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X, y_new, test_size=0.3, random_state=0, stratify=y_new)
```

```python
#making our model with the random forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_depth=2, random_state=0)
clf = model.fit(X_train, Y_train)
```

```python
#identifying what features that random forest determined were important
print(model.feature_importances_)
```

```python
#predicting Y from the X test data
clfpred_test = clf.predict(X_test)
print(clfpred_test)
```

```python
#getting our classification report to see the precision and recall
from sklearn.metrics import classification_report
print(classification_report(Y_test, clfpred_test))
```

```python
#getting our confusion matrix so we can get the senstivity and specificity of our data
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, clfpred_test, labels=[0, 1]))
```

```python
#getting our accuracy score for our data
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, clfpred_test))
```

```python
#QC measures
#first up: max_depth, going up to 20 potential splits because we know this tree has to be big

from sklearn.metrics import roc_curve
from sklearn.metrics import auc

max_depths = np.linspace(1, 20, 20, endpoint=True)

# store the results in array 
train_results = []
test_results = []

for max_depth in max_depths:
    
    dt = RandomForestClassifier(max_depth=max_depth)
# Fit the model with training data, you need both X and Y for fitting the model
    dt.fit(X_train, Y_train)
    
# Predict the training data
    train_pred = dt.predict(X_train)
    
 # Get the AUC score  for training data
    false_positive_rate, true_positive_rate, thresholds = roc_curve(Y_train, train_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
   
    # Add auc score to previous train results
    train_results.append(roc_auc)
    
  # Predict the testing data  

    test_pred = dt.predict(X_test)
    
  # get the AUC score for testing data  
    
    false_positive_rate, true_positive_rate, thresholds = roc_curve(Y_test, test_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    
   # Add auc score to previous test results
    test_results.append(roc_auc)
```

```python
#making the graph pretty and putting our info from the previous cells into a pandas dataframe for easier graphing
import seaborn as sns
sns.set(style="darkgrid")

d = {'train': train_results, 'test': test_results}

pdAUC = pd.DataFrame(d)

```

```python
#plotting our max_depth
import matplotlib.pyplot as plt

fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)

ax.plot(max_depths,pdAUC.train,c='b',label='Train',fillstyle='none')
ax.plot(max_depths,pdAUC.test,c='g',label='Test')

plt.legend(loc=2)
plt.draw()
#looks like our model is pretty good with the testing set, starts to over fit at around a depth of 10
```
